```python
##  1. Limpieza de Datos (1_clean.ipynb)

# Importar paquetes
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. Carga de datos ---
df = pd.read_csv('./data/titanic.csv')

# --- 2. Informaci贸n b谩sica y valores faltantes ---
print(f"Dimensiones iniciales: {df.shape}")
print("\nInformaci贸n de tipos y nulos:")
df.info()

# Mapeo de calor de valores nulos (opcional, pero incluido en el proceso)
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Mapa de Calor de Valores Nulos')
plt.show()

# Porcentaje de valores nulos (Para Cabin, Age, Embarked)
print("\nPorcentaje de Valores Faltantes:")
missing_values_percentage = df.isnull().mean() * 100
print(missing_values_percentage.round(2).sort_values(ascending=False))

# --- 3. Correcci贸n de valores faltantes ---

# Columna Cabin: Eliminar por exceso de nulos (77.10%)
df.drop('Cabin', axis=1, inplace=True)

# Columna Age: Imputar con la mediana (m谩s robusta que la media)
median_age = df['Age'].median()
df['Age'].fillna(median_age, inplace=True)
print(f"\nEdad imputada con la mediana: {median_age}")

# Columna Embarked: Imputar con la moda (pocos nulos, variable categ贸rica)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Verificaci贸n final de nulos
print("\nVerificaci贸n final de nulos:")
print(df.isnull().sum())

# --- 4. Eliminaci贸n de variables no 煤tiles ---

# Eliminar identificadores y texto sin procesar
df.drop('Name', axis=1, inplace=True)
df.drop('Ticket', axis=1, inplace=True)
df.drop('PassengerId', axis=1, inplace=True)

print(f"\nDespu茅s de limpiar df: {df.shape[0]} filas y {df.shape[1]} columnas.")

# --- 5. Guardar datos limpios (Checkpoint) ---
df.to_csv('./data/titanic_clean.csv', index=False)
print("\nDatos limpios guardados en: ./data/titanic_clean.csv")
```
