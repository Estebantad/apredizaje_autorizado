```python
## 游늵 2. An치lisis Exploratorio de Datos (2_EDA.ipynb)

# Importar paquetes
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Colores para las gr치ficas
COLOR_NO_SURVIVED = '#240750'
COLOR_SURVIVED = '#344C64'

# --- 1. Carga de datos limpios ---
df = pd.read_csv('./data/titanic_clean.csv')

# Visualizaci칩n inicial
print("Muestra de datos limpios:")
print(df.sample(4))
print("-" * 30)

# --- 2. An치lisis de Supervivencia (Survived) ---

# Conteo de Supervivientes y Fallecidos
print("Conteo de Supervivencia:")
print(df['Survived'].value_counts())

# Gr치fica de barras de Supervivencia
plt.figure(figsize=(6, 4))
sns.barplot(x=df['Survived'].value_counts().index, y=df['Survived'].value_counts(), palette=[COLOR_NO_SURVIVED, COLOR_SURVIVED])
plt.title('Conteo de Supervivencia (0: No, 1: S칤)')
plt.xlabel('Sobrevivi칩')
plt.ylabel('Conteo')
plt.xticks([0, 1], ['No', 'S칤'])
plt.show()

# --- 3. Supervivencia por Variables Categ칩ricas ---

# Gr치fica: Supervivencia por Clase (Pclass)
plt.figure(figsize=(8, 5))
sns.countplot(df, x='Pclass', hue='Survived', palette=[COLOR_NO_SURVIVED, COLOR_SURVIVED]) 
plt.legend(title='Supervivencia', labels=['No sobrevivi칩', 'S칤 sobrevivi칩'])
plt.title('Supervivencia por Clase de Pasajero')
plt.show()

# Gr치fica: Supervivencia por G칠nero (Sex)
plt.figure(figsize=(8, 5))
sns.countplot(df, x='Sex', hue='Survived', palette=[COLOR_NO_SURVIVED, COLOR_SURVIVED])
plt.legend(title='Supervivencia', labels=['No sobrevivi칩', 'S칤 sobrevivi칩'])
plt.title('Supervivencia por G칠nero')
plt.show()

# Gr치fica: Supervivencia por Puerto de Embarque (Embarked)
plt.figure(figsize=(8, 5))
sns.countplot(df, x='Embarked', hue='Survived', palette=[COLOR_NO_SURVIVED, COLOR_SURVIVED]) 
plt.legend(title='Supervivencia', labels=['No sobrevivi칩', 'S칤 sobrevivi칩'])
plt.title('Supervivencia por Puerto de Embarque')
plt.show()

# --- 4. An치lisis de Distribuci칩n de Edad (Age) ---

# Gr치fica: Distribuci칩n de Edades con Media, Mediana y Moda
plt.figure(figsize=(10, 6))
sns.histplot(df, x='Age', kde=True)
plt.title('Distribuci칩n de Edades')
plt.axvline(df['Age'].mean(), color='red', linestyle='--', label=f"Media: {df['Age'].mean():.2f}")
plt.axvline(df['Age'].median(), color='green', linestyle='-', label=f"Mediana: {df['Age'].median():.2f}")
plt.axvline(df['Age'].mode()[0], color='black', linestyle=':', label=f"Moda: {df['Age'].mode()[0]:.2f}")
plt.legend()
plt.show()

print(f"\nEstad칤sticas de Edad: Media={df['Age'].mean():.2f}, Mediana={df['Age'].median():.2f}, Moda={df['Age'].mode()[0]:.2f}")

# Gr치fica: Boxplot de Edades para Outliers
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Age'])
plt.title('Boxplot de Edad para Detecci칩n de Outliers')
plt.show()


# --- 5. Matriz de Correlaci칩n (Variables Num칠ricas) ---

# Filtrar solo variables num칠ricas relevantes para correlaci칩n
numeric_features = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

plt.figure(figsize=(8, 6))
sns.heatmap(df[numeric_features].corr(), 
            annot=True, 
            cbar=True,
            cmap='coolwarm',
            cbar_kws={'label': 'Nivel de correlaci칩n'})
plt.title('Correlaci칩n de variables num칠ricas')
plt.show()

# --- 6. Tarea Adicional: Supervivencia por SibSp ---

# Calcular el porcentaje de supervivencia por SibSp
survival_rate_sibsp = df.groupby('SibSp')['Survived'].mean().reset_index()

# Gr치fica de tasa de supervivencia por SibSp
plt.figure(figsize=(8, 5))
sns.barplot(x='SibSp', y='Survived', data=survival_rate_sibsp, palette='viridis')
plt.title('Tasa de Supervivencia basada en SibSp (Hermanos/Esposos)')
plt.ylabel('Tasa de Supervivencia')
plt.xlabel('N칰mero de Hermanos/Esposos (SibSp)')
plt.show()
print("\nPorcentaje de Supervivencia por SibSp:")
print(survival_rate_sibsp)
```
